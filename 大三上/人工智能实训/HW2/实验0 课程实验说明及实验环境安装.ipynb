# 实验0 课程实验说明及实验环境安装

# 1 课程实验说明
《多智能体及强化学习》一课共有六次实验，前四次实验主要是以jupyter notebook（下文称“实验手册”）的方式进行。

在这四次实验中，每一次实验都有一个这样的notebook文件，其中包含对实验所需算法和强化学习**环境**（环境，是强化学习的一个专业术语，不是指开发环境）的简略讲解，以及算法和环境对应的代码实现，在每个实验手册中，都会有若干个问题待同学们作答，全部问题作答完毕后将文件名改为`姓名+学号+第x次实验`进行提交，该文件视作每次实验的作业。

jupyter notebook的注意事项：

在回答程序类题目时，同学们可以充分利用实验手册中已经定义的函数和`rl_utils.py`文件中定义的函数，减少代码的重复编写。

已运行的代码块中的内容都会保存在jupyter notebook的内核中，每次运行新代码时，需要留意之前运行的代码是否对某些关键变量进行了修改，如果需要清除所有的记录，可以**重启内核**






# 2 实验环境安装

## 2.1 安装Anaconda或Miniconda以及jupyuter notebook

> 已经有这些环境的同学可以跳过这一步（通常Anaconda会自带jupyter notebook）

步骤如下：
- 先安装miniconda，并设置好环境变量，对conda进行换源（参考链接1）
- 设置jupyter notebook的默认打开路径（参考链接2）
- 创建一个虚拟环境`jupyter`，并安装jupyter notebook（这样做的好处是把jupyter notebook跟其他虚拟环境相隔开）（参考链接1）

安装miniconda、jupyter、虚拟环境参考教程：https://zhuanlan.zhihu.com/p/449750184

设置jupyter notebook的默认打开路径：https://blog.csdn.net/weixin_43816759/article/details/119244878

## 2.2 创建实验所需的虚拟环境
 
步骤如下：
- 在miniconda创建一个虚拟环境，名字同学们选容易记忆的就可以（如`myrl`），python的版本选择`python=3.7.12`（或3.7.16）


安装miniconda、jupyter、虚拟环境参考教程：https://zhuanlan.zhihu.com/p/449750184

> 注意，目前还不能用jupyter notebook打开我们创建的虚拟环境，因为我们还未在我们的环境中安装jupyter内核

## 2.3 安装各种包

以下所有的安装都是用pip安装，并使用清华源`https://pypi.tuna.tsinghua.edu.cn/simple`(在安装指令后面加 `-i https://pypi.tuna.tsinghua.edu.cn/simple`即可)，请按以下顺序进行安装，否则会出现版本冲突的情况。


- 安装 numpy 1.21.6 `pip install numpy==1.21.6 -i https://pypi.tuna.tsinghua.edu.cn/simple` （格式下同）
- 安装 matplotlib 3.2.2
- 安装 jupyter 内核 `pip install ipykernel`
- 激活内核 `python -m ipykernel install --user --name=你的虚拟环境的名字`
- 安装 gym 0.19.0
- 安装 tqdm
- 安装 torch（cpu版） `pip install torch==1.10.0+cpu torchvision==0.11.0+cpu torchaudio==0.10.0 -f https://download.pytorch.org/whl/torch_stable.html`
- 安装 pyglet 1.5.0


# 3 可能出现的一些问题

> 若严格按照上文的安装方式，一般不会发生以下问题

## 3.1 切换kernel报错

在jupyter中切换myrl的kernel时**zmq**报错：

`ImportError: DLL load failed: 找不到指定的模块。`

原因是zmq版本过新，需要重新安装旧版本

`pip install pyzmq==23.0.0`

如果先安装matplotlib再安装ipykernel则不会发生此问题


## 3.2 如果numpy版本是1.21.5
重装1.21.6的numpy

## 3.3 torch超时报错

如果是安装某个包（如pillow）的时候一直超时报错，可以先安装这个包（注意好版本）

## 3.4 matplotlib报错

matplotlib报错 `dll load failed: 找不到指定的模块`。

原因是与numpy版本不兼容，需要更换matplotlib版本，按照上面的推荐版本安装应该不会有这个问题
